# Age

```python
ax = snd.countplot(x = "Age Group",data=df,hue="Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
## Total Amount Vs Age Group
sales_age = df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending = False)

sns.barplot(data=sales_age,x = "Age Group",y="Amount",hue="Age Group")
```

# States

```python
## totals Number of orders from 10 states
```

```python
sales_states = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})## For Changing width and height of barplot
sns.barplot(data=sales_states,x = "State" ,y = "Orders")
```

```python
## Total amount/sales from top 10 set

sales_states = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})## For Changing width and height of barplot
sns.barplot(data=sales_states,x = "State" ,y = "Amount")
```

from above graph  we can see that most of the orders are from uttarpradesh ,maharastra and karnataka respectively

# Maretial_Status

```python
ax = snd.countplot(x = "Marital_Status",data=df)
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
## Here we checking on the basis of Marital_Status
sales_states = df.groupby(["Marital_Status",'Gender'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending = False).head(10)
sns.set(rc={'figure.figsize':(6,5)})## For Changing width and height of barplot
sns.barplot(data=sales_states,x = "Marital_Status" ,y = "Amount", hue="Gender")
```

*from above graph we can see that most of the buyer are married(women) and they have high purchasing power

# Occupation

```python
yx = snd.countplot(x = "Occupation",data=df)
sns.set(rc={'figure.figsize':(40,5)})
for bars in yx.containers:
    yx.bar_label(bars)
```

```python
# Occupation vs Ammount

sales_states = df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})## For Changing width and height of barplot
sns.barplot(data=sales_states,x = "Occupation" ,y = "Amount")
```

from above graph we can see that the most of the buyers are working in IT, Aviation And Helthcare

# Product_Category

```python
yx = snd.countplot(x = "Product_Category",data=df)
sns.set(rc={'figure.figsize':(20,5)})
for bars in yx.containers:
    yx.bar_label(bars)
```

```python
## here also i categoriged based on Product_Category and Amount
sales_states = df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})## For Changing width and height of barplot
sns.barplot(data=sales_states,x = "Product_Category" ,y = "Amount")
```

from above graph we can see thats most of the product sold are from food, clothing, electronics

```python
sales_states = df.groupby(["Product_Category"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})## For Changing width and height of barplot
sns.barplot(data=sales_states,x = "Product_Category" ,y = "Orders")
```
